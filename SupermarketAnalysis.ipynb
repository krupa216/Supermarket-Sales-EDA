```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("SuperMarket Analysis.csv")

# Create Total column if not present
if 'Total' not in df.columns:
    df['Total'] = df['Unit price'] * df['Quantity']

# Convert Date column
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    num_df = df[['Unit price', 'Quantity', 'Rating', 'Total']]

# Compute correlation matrix
corr = num_df.corr()

# Plot correlation heatmap using matplotlib
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap (Matplotlib)")
plt.show()

# 1. BAR CHART  Sales by Branch

branch_sales = df.groupby('Branch')['Total'].sum()
plt.bar(branch_sales.index, branch_sales.values)
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.title("Total Sales by Branch")
plt.show()



# 2. BAR CHART  Sales by Product Line

product_sales = df.groupby('Product line')['Total'].sum()
plt.bar(product_sales.index, product_sales.values)
plt.xticks(rotation=45)
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.title("Sales by Product Line")
plt.show()

# 3. BAR CHART  Sales by Gender

gender_sales = df.groupby('Gender')['Total'].sum()
plt.bar(gender_sales.index, gender_sales.values)
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.title("Gender-wise Sales")
plt.show()

# 4. PIE CHART  Payment Method

payment_sales = df.groupby('Payment')['Total'].sum()
plt.pie(payment_sales, labels=payment_sales.index, autopct='%1.1f%%')
plt.title("Sales Distribution by Payment Method")
plt.show()


# 5. LINE PLOT  Daily Sales Trend

if 'Date' in df.columns:
    daily_sales = df.groupby('Date')['Total'].sum()
    plt.plot(daily_sales)
    plt.xlabel("Date")
    plt.ylabel("Total Sales")
    plt.title("Daily Sales Trend")
    plt.show()

# 6. HISTOGRAM  Rating Distribution

plt.hist(df['Rating'], bins=10)
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Customer Rating Distribution")
plt.show()


# 7. SCATTER PLOT  Quantity vs Total

plt.scatter(df['Quantity'], df['Total'])
plt.xlabel("Quantity")
plt.ylabel("Total Sales")
plt.title("Quantity vs Total Sales")
plt.show()



# 8. BOX PLOT  Total Sales (Outliers)

plt.boxplot(df['Total'])
plt.ylabel("Total Sales")
plt.title("Outlier Detection in Total Sales")
plt.show()



# 9. LINE PLOT  Daily Sales Trend

if 'Date' in df.columns:
    daily_sales = df.groupby('Date')['Total'].sum()
    plt.plot(daily_sales)
    plt.xlabel("Date")
    plt.ylabel("Total Sales")
    plt.title("Daily Sales Trend")
    plt.show()



# 10. BAR CHART  Average Rating by Branch

rating_branch = df.groupby('Branch')['Rating'].mean()
plt.bar(rating_branch.index, rating_branch.values)
plt.xlabel("Branch")
plt.ylabel("Average Rating")
plt.title("Average Customer Rating by Branch")
plt.show()


# 11. BAR CHART  Quantity Sold by Product Line

qty_product = df.groupby('Product line')['Quantity'].sum()
plt.bar(qty_product.index, qty_product.values)
plt.xticks(rotation=45)
plt.xlabel("Product Line")
plt.ylabel("Quantity Sold")
plt.title("Quantity Sold by Product Line")
plt.show()
```
